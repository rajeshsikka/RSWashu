### NASA Mars News

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from selenium import webdriver 
from splinter import Browser
import pandas as pd
```

```python
mars_news = {}
news_title = []
news_description = []
```

```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/'
```

```python
# Retrieve page with the requests module
page = requests.get(url)
page
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
#soup = BeautifulSoup(page.content, 'html5lib')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# capture the content title from the Mars website 
results = soup.find_all("div", class_="content_title")
#results_desc = soup.find_all('div', class_="image_and_description_container")
# results = soup.select_one_all('div', class_="ul.item_list li.slide")
#results = soup.find_all('div', class_="article_teaser_body")
print(results)
```

```python
# Loop through returned results

for result in results:
    #desc = result.find('div', class_ = 'rollover_description').text
    #inner_desc = result.find('div', class_ ="rollover_description_inner").text
    #content_title = result.find('a href', class_ ="content_title").text
    #text_0 = link.findNextSibling(text=True)
    #link = result.a
    #rint(link.text.strip())
    
    #news_title = result.a
    #print(news_title.text.strip())
    
    
    
    title = result.a
    title = title.text.strip()
    news_title.append(title)

print(news_title[0])
print(news_title[1])
print(news_title[2])
print(news_title[3])
print(news_title[4])
print(news_title[5])
```

```python
# capture the description from the Mars website 
results = soup.find_all("div", class_="rollover_description_inner")
print(results)
```

```python
# Loop through returned results

for result in results:
    #news_description = result.get_text().strip()
    desc = result.get_text().strip()
    news_description.append(desc)
    
print(news_description[0])
print(news_description[1])
print(news_description[2])
print(news_description[3])
print(news_description[4])
print(news_description[5])
```

```python
for i in range(6):

    print("News Title", i + 1, ":", news_title[i])
    print("Description:", news_description[i])     
    mars_news["news_title"] = news_title[i]
    mars_news["news_description"] = news_description[i]
    
    print(mars_news)
```

### Mars Hemispheres

```python
browser = Browser('chrome', headless=False)
mars_hemispheres_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
```

```python
browser.visit(mars_hemispheres_url) 
mars_hemispheres_html = browser.html
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
mars_hemispheres_soup = BeautifulSoup(mars_hemispheres_html, 'html.parser')
```

```python
print(mars_hemispheres_soup.prettify())
```

```python
hemispheres_url_info = []

products = mars_hemispheres_soup.find('div', class_='result-list')
hemispheres = products.find_all('div', class_='item') 

for hemisphere in hemispheres: 
    title = hemisphere.find('div', class_='description')
    title_text = title.a.text
    title_text = title_text.replace(' Enhanced', '')
    
    browser.click_link_by_partial_text(title_text)                          
    
    mars_hemispheres_html  = browser.html                                    
    mars_hemispheres_soup = BeautifulSoup(mars_hemispheres_html, 'html.parser')                                 
    
    image = mars_hemispheres_soup.find('div', class_='downloads').find('ul').find('li')  
    img_url = image.a['href']
    
    hemispheres_url_info.append({'title': title_text, 'img_url': img_url}) 
    
    browser.click_link_by_partial_text('Back')                               
```
